```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('errors_time.dat')
time=data[:,1]
error=data[:,0]*0.01
algs=['decision tree', 'Random Forest_100', 'Random Forest_300', 
      'Logistic Regression', 'Naive Bayes', 'Adaboost_J48_20',
      'Adaboost_J48_100', 'LogitBoost_10 stumps', 
      'LogitBoost_100 stumps', 'LogitBoost_100 stumps_95',
      'LogitBoost_25 M5P', 'SVM']
fig, ax=plt.subplots()
#ax.scatter(time, error, 'ro')
ax.scatter(np.log(time), error)
for i, txt in enumerate(algs):
    ax.annotate(txt, (np.log(time[i]), error[i]), fontsize=9)
#ax.set_xscale('log')
plt.xlabel('logged time')
plt.ylabel('error')
plt.show()
```
