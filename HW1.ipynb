```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
## Problem A ##
X_train = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_train.data")
Y_train = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_train.labels")
X_test = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_valid.data")
Y_test = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_valid.labels")
X_train
```

```python
error_train =[]
error_test =[]
for i in range(1,13,1):
  clf = DecisionTreeClassifier(criterion="entropy", max_depth=i)
  # Train Decision Tree Classifer
  clf = clf.fit(X_train,Y_train)

  #Predict the response for test dataset
  Y_pred_train = clf.predict(X_train)
  Y_pred_test=clf.predict(X_test)
  error_t = 1-metrics.accuracy_score(Y_train, Y_pred_train)
  error_T = 1-metrics.accuracy_score(Y_test, Y_pred_test)
  error_train.append(error_t)
  error_test.append(error_T)
```

```python
error_train
```

```python
plt.plot(range(1,13,1), error_train, label = 'Training error' )
plt.plot(range(1,13,1), error_test, label = 'Test error' )
plt.grid()
plt.legend()
```

```python
## Problem C ##
X_train = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_train.data")
Y_train = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_train.labels")
X_test = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_valid.data")
Y_test = np.loadtxt(r"C:\Users\18800\Desktop\20 Fall\STA 5635 Applied Machine Learning\Data\MADELON\madelon_valid.labels")
X_train
```

```python
kTrees =[3,10,30,100,300]
error_train =[]
error_test =[]
for kTree in kTrees:
  clf = RandomForestClassifier(n_estimators=kTree,criterion="entropy", max_features= 'sqrt')
  # Train Random Forest Classifier
  clf = clf.fit(X_train,Y_train)

  #Predict the response for test dataset
  Y_pred_train = clf.predict(X_train)
  Y_pred_test=clf.predict(X_test)
  error_t = 1-metrics.accuracy_score(Y_train, Y_pred_train)
  error_T = 1-metrics.accuracy_score(Y_test, Y_pred_test)
  error_train.append(error_t)
  error_test.append(error_T)
```

```python
plt.plot(kTrees,error_train, label = 'training error' )
plt.plot(kTrees, error_test, label = 'test error')
plt.xlabel('number of trees')
plt.ylabel('error rate')
plt.grid()
plt.legend()
```
